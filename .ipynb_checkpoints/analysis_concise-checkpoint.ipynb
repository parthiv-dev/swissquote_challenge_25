# Weighting of Thematic Portfolio "Vital Infrastructure"

This is a short concise python notebook to intuitively explain the step-by-step process of our portfolio weighting schema.

### 1.Import necessary packages

```python
import pandas as pd
import numpy as np
```

### 2. Import & clean data 

```python
df = pd.read_csv('data_closing_prices.csv')

# Print data
df
```

```python
df.ffill(inplace=True)

# Print cleaned data
df
```

### 3. Calculating daily price changes

```python
# Calculate the fractional difference between consecutive rows
df_diff = df.pct_change()

# Drop the first row with NaN values
df_diff = df_diff.dropna()

# Print the result
df_diff
```

### 4. Calculating final weighting

```python
# Define a new dataframe to store the volatility and inverse volatility
df_mini = pd.DataFrame()

# Calculate the volatility for each column in the old dataframe
df_mini['Volatility'] = df_diff.std()

# Calculate the inverse volatility
df_mini['Inverse Volatility'] = 1 / df_mini['Volatility']

# Calculate the sum of the inverse volatilities
sum_inverse_volatility = df_mini['Inverse Volatility'].sum()

# Calculate the volatility weighting for each stock based on the inverse volatility and sum of all inverse volatilities
df_mini['Volatility Weighting'] = df_mini['Inverse Volatility'] / sum_inverse_volatility

# Calculate the volatility weighting in percentage (multiply by 100 & round to 2 decimal places)
df_mini['Volatility Weighting (%)'] = (df_mini['Volatility Weighting'] * 100).round(2)

# Print the result
df_mini
```

```python
# Print number of stocks:
print(f"Number of stocks: {len(df_mini)}")
# Print name of stock with lowest and highest volatility
print(f"Lowest weighting (%): {df_mini['Volatility Weighting (%)'].idxmin()} ({df_mini['Volatility Weighting (%)'].min()})")
print(f"Highest weighting (%): {df_mini['Volatility Weighting (%)'].idxmax()} ({df_mini['Volatility Weighting (%)'].max()})")
# Print sum of final weighting to verify correctness
print(f"Sum of final weighting: {df_mini['Volatility Weighting'].sum()}")
```
